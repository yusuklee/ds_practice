### 🔗 Association Rules

연관 규칙(Association Rules): **A가 발생하면 B도 같이 발생하는 패턴을 발견하는 것**

FP-Growth, Apriori 알고리즘으로 연관 규칙을 찾을 수 있으며, 그렇게 찾은 규칙이 **얼마나 의미 있는지**를 판단하기 위해 아래와 같은 지표를 사용함:

- **📌 지지도(Support)**: 전체 Transaction 중 특정 아이템 세트가 포함된 비율  
  - 예: 빵이 1000개의 트랜젝션(거래) 중 125건에서 함께 등장했다면,  
    \[
    Support(빵) = 125 / 1000 = 12.5\%

    Support(우유) = 150 / 1000 = 15.0\%

    Support(빵 ∩ 우유) = 100 / 1000 = 10.0\%
    \]

- **📌 신뢰도(Confidence)**: A가 포함된 거래 중 B도 함께 포함될 확률 (A → B)  
    \[
    Confidence(빵 → 우유) = Support(빵 ∩ 우유) / Support(빵) = 100/125 = 80\%
    \]

    해석: 빵을 구매한 고객 중 80%가 우유도 구매했다는 뜻

- **📌 향상도(Lift)**: A가 존재할 때 B가 나타날 확률이 우연히 발생할 확률보다 얼마나 높은지 측정  
    \[
    Lift(빵 → 우유) = Confidence(빵 → 우유) / Support(우유) = 0.8/0.15 = 5.33
    \]

    해석: 빵을 구매한 사람이 우유를 구매할 확률이, 전체 고객이 우유를 살 확률보다 5.33배 더 높다는 의미.

  - \( Lift > 1 \): 연관성이 강함  
  - \( Lift = 1 \): 독립적  
  - \( Lift < 1 \): 연관성이 약함
