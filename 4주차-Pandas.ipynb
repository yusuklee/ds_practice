## ✅ ***.iloc vs .loc 비교***

$$
\begin{array}{|c|c|c|}
\hline
\textbf{항목} & \texttt{.iloc} & \texttt{.loc} \\
\hline
\textbf{의미} & \text{Integer-location (정수 위치 기반)} & \text{Label-location (이름 기반)} \\
\hline
\textbf{기준} & \text{행 또는 열의 위치 인덱스 (0,1,2,...)} & \text{행 또는 열의 라벨 이름 (예: 'userID')} \\
\hline
\textbf{스타일} & \text{NumPy 스타일 (순서 기반)} & \text{Pandas 고유 스타일 (레이블 기반)} \\
\hline
\textbf{접근 예시} & \texttt{df.iloc[0]} & \texttt{df.loc[1001]} \\
\hline
\textbf{사용 목적} & \text{순서 기반 인덱싱, 위치 기반 추출} & \text{이름 기반 접근, 의미 있는 라벨로 조회} \\
\hline
\end{array}
$$

## ✅ ***Group by***

✔ 같은 값을 하나로 묶어 통계 혹은 집계 결과를 얻기위해 활용 \
✔ 묶은 이후 Aggregating function으로 다음과 같은 함수 활용 가능
- mean()
- sum()
- size()
- count()
- std()
- min()
- max()

```python
df_new = pd.DataFrame({
    'city': ['부산', '부산', '부산', '부산', '서울', '서울', '서울'],
    'fruits': ['apple', 'orange', 'banana', 'banana', 'apple', 'apple', 'banana'],
    'price': [100, 200, 250, 300, 150, 200, 400],
    'quantity': [1, 2, 3, 4, 5, 6, 7]
})
df_new
```

#### Size() vs count()
- size는 NaN value를 포함하여 세지만, count는 그렇지 않음
- 본질적으로 count는 값의 갯수를, size는 오브젝트의 갯수를 세기때문에 발생하는 차이임   
- 또한 size는 Pandas.Series를, count는 Pandas.DataFrame을 return함
