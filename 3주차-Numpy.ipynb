# ✅ ***Reshape이 빠른 이유***

- 배열의 데이터를 **실제로 복사하는 게 아니라**, 다른 "모양"으로 **바라보는 방식만** 바꾸기 때문

### ▶ **1. NumPy 배열의 구조**
- NumPy 배열은 연속된 메모리 공간에 데이터를 저장함
- 예를 들어, 1D or 2D or 3D 배열의 값들은 물리적으로 메모리 안에 연속적으로 저장되어 있음
- 이때 배열을 reshape하면, 실제로 메모리 안에 저장된 **"데이터를 옮기거나 복사하지 않고"**, 다른 모양으로 바라보는 방식(view)만 변경함

### ▶ **2. Reshape의 동작 방식**
- Reshape은 배열의 "모양"을 변경하는 함수 (메모리 안에 저장된 데이터를 옮기거나 복사하는게 아니라)
- 예를 들어, 원래 1차원 배열이 있으면, 이를 2차원 배열로 바꾸거나, 그 반대로 변환할 수 있음
- 하지만 중요한 점은, **"데이터 자체는 그대로 유지된다는 것"**

### ▶ **3. 왜 빠를까?**
- Reshape은 배열의 모양을 바꿀 때 데이터를 복사하지 않고, **기존 배열의 데이터를 재사용하기 때문에** 매우 빠름
- 배열의 메모리 위치를 "보기(view)"만 바꾸는 방식으로 동작해서, 불필요한 메모리 복사나 이동이 없으므로 매우 효율적임

### ▶ **4. 결론**
- Reshape은 메모리 공간에서 **실제 데이터가 복사되거나 이동하지 않고, 데이터가 보이는 "모양만 바꾸기 때문에"** 속도가 매우 빠름

## **Numpy indexing and slicing**

### **Indexing**
- Numpy 배열의 요소는 index배열로 참조할 수 있으며 참조 후 값의 수정이 가능함  
- N차원 배열은 N개의 index로 참조 할 수 있음

```python
arr, arr34
```

### **Slicing**
- ' : (콜론)' 을 활용하여 여러 개의 요소를 참조하는 슬라이싱을 사용  
- axis별로 범위를 지정해야 하며, (from index) : (to index) 형태를 가짐  
- 이때 from index와 to index는 생략 가능하며, 생략 시 각각 '처음부터/끝까지'의 의미를 가짐  
- 각 index는 음수를 지정가능하며, 이는 반대방향을 의미 (-1은 마지막 index)

```python
arr = np.arange(1, 13, 1)
arr34 = arr.reshape(3,4)
arr, arr34
```

```python
"""Slicing"""
# arr 배열에서 4번째와 5번째 요소를 참조
# 3:5는 3번 인덱스부터 4번 인덱스까지 - [3:5)
arr[3:5]
```

```python
"""Slicing with multi index"""
# 두 번째 행의 두 번째와 세 번째 열을 참조
arr34[1:2, 1:3]
```

```python
# 첫 번째와 세 번째 행을 모두 선택
arr34[[0,2], :]
```

```python
# 모든 행에서 두 번째와 세 번째 열을 선택
arr34[:, 1:3]
```

```python
# 첫 번째와 마지막 행/열을 제외한 부분을 선택
arr34[1:-1, 1:-1]
```

```python
arr34
```

```python
"""Fancy indexing"""
# Fancy 인덱싱은 배열을 특정한 방식으로 선택하는 고급 인덱싱 기법
# (원하는 요소를 자유롭게 선택할 수 있음)
# 예: 각 행에서 열 index [2, 3]와 [1, 2]를 선택하는 방식

arr34[:, [[2, 3],[1,2]]] #Fancy indexing이라고 함
```

### **Boolean numpy indexing and masking**

- Numpy는 **shape이 다른 두 배열에 대한 연산을 지원**함  
- 작은 배열을 큰 배열 크기에 맞게 자동으로 확장해서 연산


```python
# NumPy broadcasting: NumPy가 연산이 빠른 이유 중 하나
# NumPy의 가장 큰 장점 중 하나가 '반복문 없이 배열 전체에 연산을 한 번에 적용. 즉, 벡터화'를 할 수 있는 건데,
# 이걸 가능하게 해주는 게 바로 broadcasting
# Broadcasting이 없으면? : 모든 shape을 맞추기 위해 reshape()으로 강제로 맞춰야 함
```

```python
arr = np.arange(0, 12).reshape(3,4)
arr
```

```python
# 여기서 브로드캐스팅이 일어남
# 3이 3X4 차원의 배열로 확장되어 각 요소에 더해진 것

arr + 3

# 만약 python for 문으로 구현하면? : 각 행마다 for, 각 요소마다 for
```

- ✅ **Broadcasting 조건 :**\
차원(dimension)이 같은 두 배열의 각 차원의 크기(axis)를 비교해서,\
✔ **1. 같은 크기이거나,**\
✔ **2. 둘 중 하나가 1이면 가능**\
✔ *다만, 두 배열의 차원 수(dim)가 다른데 작은 배열의 차원이 1이면 shape 앞에 1을 자동으로 추가(차원 수 자동 확장)*

- 예시) \
✔ (4, 3) + (1, 3) \
➡ 가능! (1이므로 확장됨)\
✔ (2, 3) + (3,)    \
➡ 가능! (작은 배열의 차원이 1이면 shape 앞에 1을 자동으로 추가)\
✔ (5, 4, 3) + (1, 4, 3)  \
➡ 가능! (1이므로 확장됨)\
✔ (2, 3, 4) + (2, 3, 2)  \
➡ 불가능! (마지막 차원 수가 다름)

```python
# 이 두 배열은 서로 차원이 다르지만, broadcasting을 통해 연산 가능

arr1 = np.arange(5).reshape(1, 5)
arr2 = np.arange(5).reshape(5, 1)
arr1, arr2
```

```python
# 각각 arr1의 차원에 맞게 확장하여 arr2를 (5, 5) 형태로 만들고, 이를 더하는 연산을 수행
# 자동으로 broadcasting이 발생하기 때문

arr1 + arr2
```
